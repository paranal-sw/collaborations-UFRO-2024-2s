El método 'sample_meta' define al atributo 'target_data' que es un subconjunto de datos con el cual trabajar.
Su argumento principal es 'success_ratio' que selecciona todos los errores y los concatena con una cantidad de éxitos aleatorios proporcional al valor ingresado.
(por defecto su valor es 1, es decir, si la muestra tiene $n$ errores también lo tendrán los éxitos, si el valor es 2 se eligen $n$ errores y $2n$ éxitos.)

Si se elige el valor de target_meta = True, entonces el atributo apuntará directamente a "meta_data".

```python
Base.target_data  # success_ratio = 1
```

```python
Base.sample_meta(success_ratio=3)  # Si hay 2 errores, entonces hay 2*3= 6 éxitos (ERROR = False)
Base.target_data
```

```python
# Cuando se utiliza target_meta = True, se apuntará a la misma celda de memoria que el atributo meta_data (target_data = meta_data)
Base.sample_meta(target_meta=True)
print('target_data memory identifier: ',id(Base.target_data))
print('meta_data memory identifier: ',id(Base.meta_data))
print('Equal: ',id(Base.target_data) == id(Base.meta_data))
Base.target_data.head()
```

El método trace_tokenization devuelve una tokenización de la columna "logtext" en la traza como un objeto pandas.Series. Bajo el setting 'RegExp' se utiliza un tokenizador de expresiones regulares (por ahora es el único). Esto se realiza llamando a la clase "Custom_Tokenizer" del archivo Structure.

Puede incorporar como kwargs un diccionario con argumentos dependiente del setting.

setting = 'RegExp':
  -  'regexp' (str) : expresión regular para realizar la tokenización.
  -  'gaps'(bool) : si es True se omiten espacios en la tokenización.

```python
# Algunas expresiones regulares para probar

tokenizer_args = {'regexp': r'\s+', 'gaps' : True}  # Conservar letras y reemplazar el resto por espacios.
Trace = Base.trace_tokenization(**tokenizer_args)

print(Trace.head())

tokenizer_args = {'regexp': r'\D+', 'gaps': False}  # Eliminar solo números
Trace = Base.trace_tokenization(**tokenizer_args)
print(Trace.head())

Trace = Base.trace_tokenization()  # default  regexp = r'\b[a-zA-Z]+\b' , gaps = False
print(Trace.head())
```

Para tokenizaciones se recomienda trabajar solo en el conjunto objetivo (target_data) en lugar del meta_data completo. Transmite los mismos argumentos que trace_tokenization en un diccionario de kwargs.

El resultado es una columna nueva llamada Trace_tokens que almacena en una lista de listas la tokenización de los logs.

```python
tokenizer_args = {'regexp': r'\D+', 'gaps': False}  # Eliminar solo números
Base.target_tokenization(**tokenizer_args)
Base.target_data.head()
```

La vectorización se realiza utilizando el método 'target_vetorization' que por ahora solo utiliza un CountVectorizer, también permite la recepción de una lista de stopwords mediante el kwarg 'stop_words'.

**Pipeline** : El pipeline para entrenar y/o evaluar modelos debe adaptarse por casos, estos son señalados en el argumento 'method'.

Si 'method' es igual a "Multinomial_NB", entrenará un clasificador Naive Bayes. Se consideran los argumentos "test_size"(transmitido a un train_test_split estratificado de sklearn con semilla fija), y un diccionario 'NB_kwargs' de parámetros para aplicar al modelo.

Por otro lado si 'method' es igual a 'TextBlob' se evaluarán directamente en el objeto "TextBlob_Classifier" los tokens almacenados en target_data aplicando un muestreo estratificado de tamaño "test_size". También recibe el argumento "treshold" para transmitir al clasificador BlobText.

Las evaluaciones se realizan por métricas y son almacenadas en el atributo data del pipeline asociado al Core.

```python
Base.call_pipeline(method='Multinomial_NB', test_size=0.3, NB_kwargs={'alpha':0.1})
Base.call_pipeline(method='Multinomial_NB', test_size=0.3, NB_kwargs={'alpha':0.5})
Base.call_pipeline(method='Multinomial_NB', test_size=0.3, NB_kwargs={'alpha':0.9})
Base.call_pipeline(method='Multinomial_NB', test_size=0.3, NB_kwargs={'alpha':1.0})

tokenizer_args = {'regexp': r'\D+', 'gaps': False}
Base.target_tokenization(**tokenizer_args)  # Cambiar expresión regular
Base.target_vectorization(stop_words=stop_words)
Base.call_pipeline(method='Multinomial_NB', test_size=0.3, NB_kwargs={'alpha':0.2})
Base.call_pipeline(method='Multinomial_NB', test_size=0.3, NB_kwargs={'alpha':0.5})
Base.call_pipeline(method='Multinomial_NB', test_size=0.3, NB_kwargs={'alpha':0.5, 'fit_prior' : False})
Base.call_pipeline(method='Multinomial_NB', test_size=0.3, NB_kwargs={'alpha':1.0})
```

```python
# Resultados son representables en un DataFrame
import pandas as pd

result = pd.DataFrame(Base.pipeline.data)
result
```
