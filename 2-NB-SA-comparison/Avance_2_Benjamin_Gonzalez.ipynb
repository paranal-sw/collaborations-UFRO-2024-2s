```python
#Genero una funcion que saca todos los logs con cada trace id y los deja en una lista (Esta funcion es solo para utilizarla con el df_ints)
def listas_logtext(df):
    #Aca guardo los valores de las trace_id
    traces = df['trace_id'].unique()
    listas_log=[] #Lista vacia para acumular listas de logs de cada trace_id

    for traza in traces:
        trace = df[df['trace_id']== traza]
        log_text_trace_id = (list(trace['logtext']))
        listas_log.append(log_text_trace_id)
    
    return listas_log

trace_log= listas_logtext(df_inst)
```

```python
from collections import Counter
```

```python
#Hare una lista con conteos de los logs que se mostraron en cada trace_id
def conteo_traces_id(listas_de_logs):

    lista_logs_distintos = []
    for lista in listas_de_logs:
        conteo_elementos =Counter(lista)
        lista_logs_distintos.append(conteo_elementos)
    return lista_logs_distintos

#A la funcion de conteo de entrada le doy las listas que generaron error en los trace_id
conteo_logs = conteo_traces_id(trazas_error)
```

```python
import re

def preprocesador_1(text):
    #Se eliminan los caracteres que no sean letras del abecedario
    puntuacion = r'[,º;.:•“”¡!³¿?<>@#$%&[\](){}<>~=+\-*/|\\_^`"\']'
    text = re.sub(puntuacion, ' ', text)
    text = re.sub('[\W]+', ' ', text.lower()) 
    
    #Se eliminan las tildes de las vocales y espacios en blanco sobrantes
    text = re.sub(r'\s+', ' ', text).strip()
    
    #Tokenizar
    tokens = text.split()
    
    
    #Unir tokens nuevamente en un solo texto
    texto_preprocesado = ' '.join(tokens)
    
    return texto_preprocesado
```

```python
for i,lista in enumerate(df_nuevo_meta['trace_logs']):
    for j,texto in enumerate(lista):
        texto_limpio = preprocesador_1(texto)
        df_nuevo_meta['trace_logs'][i][j] = (df_nuevo_meta['trace_logs'][i][j]).replace(texto,texto_limpio)
```

```python
from textblob import TextBlob

# Lista de elementos en inglés
elementos = [
    "I love this product!",
    "This is okay.",
    "I hate this service.",
    "It's amazing!",
    "Not great, not terrible."
]

# Función para clasificar el sentimiento
def clasificar_sentimiento(texto):
    analisis = TextBlob(texto)
    if analisis.sentiment.polarity > 0:
        return "positivo"
    elif analisis.sentiment.polarity < 0:
        return "negativo"
    else:
        return "neutro"

# Revisar cada elemento en la lista
resultados = {elemento: clasificar_sentimiento(elemento) for elemento in elementos}

# Imprimir resultados
for elemento, sentimiento in resultados.items():
    print(f"'{elemento}' es {sentimiento}.")
```

```python
#Estos son los logs del trace_id=0 en los que voy a hacer la clasificacion de los logs
texto = df_nuevo_meta['trace_logs'][18]

# Revisar cada elemento en la lista
resultados = {elemento: clasificar_sentimiento(elemento) for elemento in texto}

# Imprimir resultados
sentimientos = []
for elemento, sentimiento in resultados.items():
    sentimientos.append((elemento,sentimiento))
    
#La lista sentimiento tiene como elementos tuplas en las que el primer elemento de la tupla es un log y el segundo elemento de la tupla es su clasificacion
print(sentimientos)
```

```python
conteo_pos=0
conteo_neg=0
conteo_neutro=0

texto_negativo=[]
texto_positivo=[]

for tupla in sentimientos:
    if tupla[1] == 'neutro':
        conteo_neutro+=1
    elif tupla[1] == 'positivo':
        texto_positivo.append(tupla[0])
        conteo_pos+=1
    elif tupla[1] == 'negativo':
        texto_negativo.append(tupla[0])
        conteo_neg+=1


print(f'Para el trace_id = {df_nuevo_meta['trace_id'][18]}')
print(f'Hubieron {conteo_pos} logs que se detectaron como positivos')
print(texto_positivo)
print('---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')

print(f'Hubieron {conteo_neg} logs que se detectaron como negativos')
print(texto_negativo)
print('------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')

print(f'Hubieron {conteo_neutro} logs que se detectaron como neutros')
```
