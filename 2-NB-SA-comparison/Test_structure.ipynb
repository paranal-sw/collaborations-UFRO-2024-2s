```python
from textblob import TextBlob

# Lista de elementos en inglés
elementos = [
    "I love this product!",
    "This is okay.",
    "I hate this service.",
    "It's amazing!",
    "Not great, not terrible."
]

# Función para clasificar el sentimiento
def clasificar_sentimiento(texto):
    analisis = TextBlob(texto)
    if analisis.sentiment.polarity > 0:
        return "positivo"
    elif analisis.sentiment.polarity < 0:
        return "negativo"
    else:
        return "neutro"

# Revisar cada elemento en la lista
resultados = {elemento: clasificar_sentimiento(elemento) for elemento in elementos}

# Imprimir resultados
for elemento, sentimiento in resultados.items():
    print(f"'{elemento}' es {sentimiento}.")
```

```python
#Estos son los logs del trace_id=0 en los que voy a hacer la clasificacion de los logs
texto = df_pioner['Trace_tokens'][0]

# Revisar cada elemento en la lista
resultados = {elemento: clasificar_sentimiento(elemento) for elemento in texto}

# Imprimir resultados
sentimientos = []
for elemento, sentimiento in resultados.items():
    sentimientos.append((elemento,sentimiento))
    
#La lista sentimiento tiene como elementos tuplas en las que el primer elemento de la tupla es un log y el segundo elemento de la tupla es su clasificacion
print(sentimientos)
```

```python
#Esto es lo que mas demora (aprox 10 min si es que buscamos en 1 mes)

clasificacion_sentimiento = []
for i in range(n):
    texto = df_pioner['Trace_tokens'][i]

    # Revisar cada elemento en la lista
    resultados = {elemento: clasificar_sentimiento(elemento) for elemento in texto}

    # Imprimir resultados
    sentimientos = []
    for elemento, sentimiento in resultados.items():
        sentimientos.append((elemento,sentimiento))

    conteo_pos=0
    conteo_neg=0
    conteo_neutro=0

    texto_negativo=[]
    texto_positivo=[]

    for tupla in sentimientos:
        if tupla[1] == 'neutro':
            conteo_neutro+=1
        elif tupla[1] == 'positivo':
            conteo_pos+=1
        elif tupla[1] == 'negativo':
            conteo_neg+=1

    if conteo_pos> conteo_neg:
        clasificacion_sentimiento.append(0)
    else:
        clasificacion_sentimiento.append(1)
```

```python
df_pioner['class_sentiment'] = clasificacion_sentimiento
```

```python
conteo_pos=0
conteo_neg=0
conteo_neutro=0

texto_negativo=[]
texto_positivo=[]

for tupla in sentimientos:
    if tupla[1] == 'neutro':
        conteo_neutro+=1
    elif tupla[1] == 'positivo':
        texto_positivo.append(tupla[0])
        conteo_pos+=1
    elif tupla[1] == 'negativo':
        texto_negativo.append(tupla[0])
        conteo_neg+=1

print(f'Para el trace_id = {df_pioner['Trace_tokens'][0]}')
print(f'Hubieron {conteo_pos} logs que se detectaron como positivos')
print(texto_positivo)
print('---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')

print(f'Hubieron {conteo_neg} logs que se detectaron como negativos')
print(texto_negativo)
print('------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')

print(f'Hubieron {conteo_neutro} logs que se detectaron como neutros')
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score

classifier = MultinomialNB()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)

accuracy = accuracy_score(y_test,y_pred)
classification_rep = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:")
print(classification_rep)
```
